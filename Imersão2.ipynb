Desafio 2.1 - Ordenar o gráfico countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure (figsize=(8,6))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = cod_compostos)
ax.set_title ('Top 5 compostos')
plt.show()
```

Desafio 2.2 - Melhorar o gráfico, fonte, etc

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure (figsize=(8,6))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = cod_compostos, color = 'mediumorchid', hatch = '/')
plt.ylabel("Quantas x foi usado", fontsize=17)
plt.xlabel("Compostos usados", fontsize=17)
ax.set_title ('Top 5 compostos', fontsize = 20)
plt.show()
```

2.3 - Plotar os histogramas com o seaborn

```python
import seaborn as sns
```

```python
plt.figure (figsize=(8,6))
sns.histplot(dados.g3, bins=50, kde=True)
plt.title("Dados de G3", fontsize = 17)
plt.ylabel("N° de experimentos feitos", fontsize=17)
plt.xlabel("Expressões gênicas", fontsize=17)

```

```python
sns.histplot(data=dados.loc[:, 'g0':'g73'].describe().T['mean'], color = 'mediumorchid', kde=True)
plt.title("Média de g0 até g73", fontsize = 17)
plt.ylabel("N° de experimentos feitos", fontsize=17)
plt.xlabel("Expressões gênicas", fontsize=17)
```

2.4 - Estudar sobre as estatísticas retornadas no .describe

count = total do conjunto
mean = o valor médio
std =  desvião padrão, indica o quanto os dados são uniformes
min = o valor mínimo dado
max = o valor máximo dado
25% = primeiro quartil
50% = mediana
75% = terceiro quartil 



2.5 - Refletir sobre a manipulação dos tamanhos das visualizações. 

Os gráficos precisam estar em um bom tamanho, se não os dados ficam distorcidos, quando formos analizarmos, não vamos conseguir especificar ou diferencia-los.
